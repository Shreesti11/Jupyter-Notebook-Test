```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

url = "https://drive.google.com/uc?id=16DeG8mBHOMjYZYsuoc1r_7ivwMZf6mOA"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# Age
plt.subplot(2, 2, 1)
sns.histplot(df['Age'], bins=20, kde=False, color='skyblue')
plt.title("Distribution of Age")

plt.tight_layout()
plt.show()
```

```python
# BMI
plt.subplot(2, 2, 2)
sns.histplot(df['BMI'], bins=20, kde=True, color='orange')
plt.title("Distribution of BMI")

plt.tight_layout()
plt.show()
```

```python
# Mental Health Days
plt.subplot(2, 2, 3)
sns.histplot(df['MentHlth'], bins=30, kde=False, color='green')
plt.title("Distribution of Mental Health Days (Past 30 days)")

plt.tight_layout()
plt.show()
```

```python
# Physical Health Days
plt.subplot(2, 2, 4)
sns.histplot(df['PhysHlth'], bins=30, kde=False, color='red')
plt.title("Distribution of Physical Health Days (Past 30 days)")

plt.tight_layout()
plt.show()
```

```python
# High Blood Pressure
plt.subplot(1, 2, 1)
sns.countplot(x='HighBP', data=df, palette='Set2')
plt.title("Prevalence of High Blood Pressure")
plt.xticks([0, 1], ['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
# High Cholesterol
plt.subplot(1, 2, 2)
sns.countplot(x='HighChol', data=df, palette='Set3')
plt.title("Prevalence of High Cholesterol")
plt.xticks([0, 1], ['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
# --- Bar Chart ---
plt.figure(figsize=(6,4))
sns.countplot(x='HeartDiseaseorAttack', data=df, palette='coolwarm')
plt.title("Distribution of Heart Disease Cases")
plt.xticks([0,1], ['No Heart Disease', 'Heart Disease'])
plt.ylabel("Count")
plt.show()
```

```python
# --- Pie Chart ---
plt.figure(figsize=(6,6))
df['HeartDiseaseorAttack'].value_counts().plot.pie(
    labels=['No Heart Disease', 'Heart Disease'],
    autopct='%1.1f%%',
    colors=['lightgreen','salmon'],
    startangle=90,
    explode=[0, 0.1]  # highlight heart disease slice
)
plt.title("Heart Disease Prevalence")
plt.ylabel("")
plt.show()
```
